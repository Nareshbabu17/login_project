```python
import re
def register():
    database = open("db.txt","r")
    username = input("Create username:")
    password = input("Create password:")
    password1 = input("Confirm password:")
    d = []
    p = []
    for i in database:
        a,b = i.split(",")
        b = b.strip()
        d.append(a)
        p.append(b)
    data = dict(zip(d,p))
    pass_ = []
    user_ = []
    for i in password:
        pass_.append(i)
    for i in username:
        user_.append(i)
    integer = [s for s in pass_ if s.isdigit()]
    lower = [s for s in pass_ if s.islower()]
    upper = [s for s in pass_ if s.isupper()]
    regex = re.compile('[@_!#$%^&*()<>?/\|}{~:]')
    regex1 = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'

    
    if password != password1:
        print("password donot match,Restart")
        register()
    elif (re.fullmatch(regex1, username))==None:
        print("InValid Email")
        register()
    else:
        if len(password)<5:
            print("Password is out of range,Restart:")
            register()
        elif len(password)>16:
            print("Password out of range,Restart")
            register()
        elif len(integer) == 0:
            print("password should contain integer")
            register()
        elif len(lower) == 0:
            print("password shoukd contain atleast one lowercase")
            register()
        elif len(upper) == 0:
            print("password should contain atleast one uppercase")
            register()
        elif(regex.search(password) == None):
            print("Password should contain integer, spl char , uppercase and lowercase")
            register()
        elif username in d:
            print("username exist")
            register()
        else:
            database = open("db.txt","a")
            database.write(username+","+password+"\n")
            print("Sucess")
        


def access():
    database = open("db.txt","r")
    username = input("Enter your  username:")
    password = input("Enter your  password:")
    
    if not len(username or password)<1:
        d = []
        p = []
        for i in database:
            a,b = i.split(",")
            b = b.strip()
            d.append(a)
            p.append(b)
        data = dict(zip(d,p))
        
        try:
            if data[username]:
                try:
                    if password == data[username]:
                        print("Login Sucess")
                        print("Hi,", username)
                    else:
                        print("password or username is incorrect")
                        
                except:
                    print("incorrect password or username")
            else:
                print("username or password doesnt exist")
                register()
                
        except:
            print("username or password doesnt exist")
            register()
    else:
        print("please enter a value")
        
        
def forgot():
    database = open("db.txt","r")
    username = input("Enter your  username:")
    d = []
    p = []
    for i in database:
        a,b = i.split(",")
        b = b.strip()
        d.append(a)
        p.append(b)
    data = dict(zip(d,p))
    print(data.get(username))
    
                    
def home(option = None):
    option = input("login | Register | forgot_password:")
    if option == "login":
        access()
    elif option == "Register":
        register()
    elif option == "forgot_password":
        forgot()
    else:
        print("please enter an option")
        
home()
```
